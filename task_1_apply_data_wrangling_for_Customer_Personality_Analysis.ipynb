```python
import pandas as pd
```

```python
df=pd.read_csv("/content/marketing_campaign.csv", sep="\t") #we used sep="\t" for column breaks /t and not ,
df
```

```python
#Data Discovery – Explore dataset structure using shape, info(), describe(), and columns inspection
print(df.shape)
print("_______________________________________________")
print(df.duplicated().sum())
print("_______________________________________________")
print(df.columns)
print("_______________________________________________")
df.info()
```

```python
df.describe()
```

```python
#Converting the column type Dt_Customer from object.. to date
df["Dt_Customer"]=pd.to_datetime(df["Dt_Customer"], format="%d-%m-%Y")
df.info()
```

```python
#Missing Data Handling – Identify missing values (isnull().sum()) and handle them using fillna() or dropna()
print(df.isnull().sum())
print("_______________________________________________")
df["Income"]=df["Income"].fillna(df["Income"].mean())
df.isnull().sum()
```

```python
#Data Type Correction – Identify columns and convert them to the correct data types (e.g., astype())
df.info()
print("_______________________________________________")
binary_cols =[
    "AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3",
    "AcceptedCmp4", "AcceptedCmp5",
    "Complain", "Response", "Kidhome", "Teenhome"]

df[binary_cols] = df[binary_cols].astype("category")
print(df.info())
#These columns were converted because they represent cases, not numbers.
```

```python
#Duplicate Removal – Clean the dataset by identifying and removing redundant entries using
#drop_duplicates()
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
print(df.duplicated().sum())
```

```python
#Basic Data Structuring – Group and summarize data using groupby() to extract initial insights
df.info()
print("_______________________________________________")
product_column =["MntWines", "MntFruits", "MntMeatProducts",
                 "MntFishProducts", "MntSweetProducts", "MntGoldProds" ]
marital_status_summary =df.groupby("Marital_Status")[product_column].mean()
print(marital_status_summary) #Average spending per product type by social status
print("_______________________________________________")
education_count = df.groupby("Education")["ID"].count()
print(education_count) #Number of person per educational level
print("_______________________________________________")
income_by_education = df.groupby("Education")["Income"].mean()
print(income_by_education) #Average income per educational category
```

```python
#Text Cleaning – Apply string operations to clean data (e.g., str.strip() and str.lower() for consistency)
df.info()
df["Education"] = df["Education"].str.strip().str.lower()
df["Marital_Status"] = df["Marital_Status"].str.strip().str.lower()
```

```python
#Final Export – Save the resulting cleaned dataset to a new CSV file
df.to_csv('Customer Personality Analysis data.csv', index=False)
```
